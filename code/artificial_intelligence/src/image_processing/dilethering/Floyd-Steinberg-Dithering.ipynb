# Floyd Steinberg Dilethering
One of the earliest Dithering algorithms is the Floyd Steinberg Dithering Algorithm. There are 2 methods to implement Floyd Steinberg Dithering in Python:
## Using Python Imaging Library (PIL):

```python
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt

imageConvert = Image.open('og_bw.png').convert(mode='1',dither=Image.FLOYDSTEINBERG)
imageConvert.save('Floyd-Steinberg-Dithering-With-PIL.png')
image = cv2.imread('Floyd-Steinberg-Dithering-With-PIL.png')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Floyd Steinberg Dithering With PIL')
```

## Floyd Steinberg Dithering from scratch:

```python
Image = cv2.imread('og_bw.png')

GrayImage = cv2.cvtColor(Image, cv2.COLOR_BGR2GRAY)
cv2.imwrite('GracyImage.PNG', GrayImage)

Height = GrayImage.shape[0]
Width = GrayImage.shape[1]

for y in range(0, Height):
    for x in range(0, Width):

        old_value = GrayImage[y, x]
        new_value = 0
        if (old_value > 128) :
            new_value = 255

        GrayImage[y, x] = new_value

        Error = old_value - new_value

        if (x<Width-1):
            NewNumber = GrayImage[y, x+1] + Error * 7 / 16
            if (NewNumber>255) : NewNumber=255
            elif (NewNumber<0) : NewNumber=0
            GrayImage[y, x+1] = NewNumber

        if (x>0 and y<Height-1):
            NewNumber = GrayImage[y+1, x-1] + Error * 3 / 16
            if (NewNumber>255) : NewNumber=255
            elif (NewNumber<0) : NewNumber=0
            GrayImage[y+1, x-1] = NewNumber

        if (y<Height-1):
            NewNumber= GrayImage[y+1, x] + Error * 5 / 16
            if (NewNumber>255) : NewNumber=255
            elif (NewNumber<0) : NewNumber=0
            GrayImage[y+1, x] = NewNumber

        if (y<Height-1 and x<Width-1):
            NewNumber = GrayImage[y+1, x+1] + Error * 1 / 16
            if (NewNumber>255) : NewNumber=255
            elif (NewNumber<0) : NewNumber=0
            GrayImage[y+1, x+1] = NewNumber
            
plt.imshow(cv2.cvtColor(GrayImage, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Floyd Steinberg Dithering With PIL')
```
