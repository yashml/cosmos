# Floyd Steinberg Dithering
One of the earliest Dithering algorithms was the Floyd Steinberg Dithering Algorithm. 

```python
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt

imageConvert = Image.open('og_bw.png').convert(mode='1',dither=Image.FLOYDSTEINBERG)
imageConvert.save('Floyd-Steinberg-Dithering-With-PIL.png')
image = cv2.imread('Floyd-Steinberg-Dithering-With-PIL.png')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Floyd Steinberg Dithering With PIL')
```
